```python
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator
import numpy as np
import xlrd
import math
from mpl_toolkits.mplot3d import Axes3D
import pandas

#Annotations######
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.set_xlabel('Strike')
ax.set_ylabel('Expiry')
ax.set_zlabel('Volatility')
ax.text2D(0.05, 0.95, "Market data plotted", transform=ax.transAxes)
fig, bx = plt.subplots(subplot_kw={"projection": "3d"})
bx.set_xlabel('Strike')
bx.set_ylabel('Expiry')
bx.set_zlabel('Volatility')
bx.text2D(0.05, 0.95, "Fitted volatility surface (SABR)", transform=bx.transAxes)
##################

# Setting time and Strike 
time = [0.25,0.5,0.75,1,2,5,10]
K =[0.5,1,1.5,1.75,2,2.25,2.5,3,3.5]
#K=[-150,-100,-50,-25,0,25,50,100,150]

KA, TIME = np.meshgrid(K, time)


#import SABR vol from cvs 
data = pandas.read_csv('5y.csv', sep = ',', header = None)
data_ = data
data = np.array([x.split(';') for x in data_[0]], dtype = np.float)

#import Market vol from cvs 
marketdata = pandas.read_csv('5ymarketdata.csv', sep = ',', header = None)
marketdata_ = marketdata
marketdata = np.array([x.split(';') for x in marketdata_[0]], dtype = np.float)

market = bx.plot_surface(KA, TIME,data , cmap=cm.coolwarm,linewidth=0, antialiased=False)
sabr = ax.plot_surface(KA, TIME, marketdata, cmap=cm.coolwarm,linewidth=0, antialiased=False)


#Plot the two surface 3D
plt.show()
```
